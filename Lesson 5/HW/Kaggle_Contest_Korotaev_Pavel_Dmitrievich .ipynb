```python
#       GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import  make_scorer

# solvers = ['newton-cg', 'lbfgs', 'sag']
solvers = ['saga', 'lbfgs', 'newton-cg']
penalty = ['l1']
c_values = [100, 10, 1.0, 0.1, 0.01]
model = LogisticRegression()

grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring=make_scorer(precision_score, average='macro'),error_score=0)
grid_result = grid_search.fit(X_train_sc, y_train)


print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector

selector = SequentialFeatureSelector(LogisticRegression( C = 10),
                                     scoring=make_scorer(precision_score, average='macro'),
                                     verbose=2, k_features=11, forward=True, n_jobs=-1)

selector.fit(X_train_sc, y_train)
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs

print('best combination (ACC: %.3f): %s\n' % (selector.k_score_, selector.k_feature_idx_))
print('all subsets:\n', selector.subsets_)
plot_sfs(selector.get_metric_dict(), kind='std_err');
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_train_sc, y_train, test_size=0.33, random_state=42)
X_train_sfs = selector.transform(X_train1)
X_test_sfs = selector.transform(X_test1)
log_reg = LogisticRegression(max_iter=10000, C = 10)
log_reg.fit(X_train_sfs, y_train1)
y_pred_N2 = log_reg.predict(X_test_sfs)
```
